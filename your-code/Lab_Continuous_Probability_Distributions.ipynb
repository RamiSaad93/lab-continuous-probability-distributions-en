## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
uniform_numbers = uniform.rvs(loc=0, scale=10, size=1, random_state=None)
# min_value = loc, max_value = loc+scale

print("Uniform:", uniform_numbers)
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
def generate_uniformal_dist(bottom, ceiling, count):
    uniform_numbers = uniform.rvs(loc=bottom, scale=ceiling - bottom, size=count, random_state=0)
    return uniform_numbers
```

```python
hundred_sample = generate_uniformal_dist(10, 15, 100)
hundred_sample
```

```python
thousand_sample = generate_uniformal_dist(10, 60, 1000)
thousand_sample
```

```python
_, charts = plt.subplots(nrows = 1, ncols=2, sharey=True)
charts[0].set_title("Sampling Distribution with n=100", fontsize=10)
charts[0].hist(hundred_sample, bins = 10)

charts[1].set_title("Sampling Distribution with n=1000", fontsize=10)
charts[1].hist(thousand_sample, bins = 10)
plt.show()
```

How are the two distributions different?

```python
from scipy.stats import norm 
```

```python
def generate_normally_dist(mean, std, count):
    normally_sample = norm.rvs(loc=mean, scale=std, size=count)
    return normally_sample
```

```python
std_1 =generate_normally_dist(10, 1, 1000)
```

```python
std_50 =generate_normally_dist(10, 50, 1000)
```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
def random_expontial(mean, size):
    random_expo = np.random.exponential(scale = mean, size = size)
    return random_expo
```

```python
size_10 = random_expontial(10, 10)
size_100 = random_expontial(10, 100)
```

```python
_, charts = plt.subplots(nrows = 1, ncols=2, sharey=True)
charts[0].set_title("Sampling Distribution with size=10", fontsize=10)
charts[0].hist(size_10, bins =100)

charts[1].set_title("Sampling Distribution with size=100", fontsize=10)
charts[1].hist(size_100, bins =100)
plt.show()
```

How are the two distributions different?

The mean changes, so the distribution changes as well.
